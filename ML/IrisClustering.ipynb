# sklearn K-means clustering on Fisher's Iris dataset

```python
%matplotlib inline

import sklearn
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# load training data
iris = sklearn.datasets.load_iris()
```

The Iris dataset has 150 instances:
* data: floats for Sepal Length, Sepal Width, Petal Length and Petal Width
* target: integer (0,1,2) representing species (Setosa, Versicolour, Virginica)

```python
# show first three rows of training instance data
print(iris.data[:3])
```

```python
# show all values for ground truth class (0,1,2)
print(iris.target)
```

```python
# clustering with 3 clusters
results = sklearn.cluster.KMeans(n_clusters=3)
results.fit(iris.data) 
predicted_labels = results.predict(iris.data)
print(predicted_labels)
```
